```python
#Sucht linknamen der letzten Bundesligisten raus

allTeams = pd.read_csv("TeamID_Team.csv")
bl_TeamsID = set(pd.read_csv("bl_mit_allspieler.csv")["team1_id"][-200:])

bl_teamslink=[]
for teamID in bl_TeamsID:
    bl_teamslink.append(allTeams['link'][teamID-1])
```

```python
import ProgMod_DsAngVer
DS_Prognose = []
for HeimMannschaftLink in bl_teamslink:
    for AuswMannschaftLink in bl_teamslink:
        if HeimMannschaftLink == AuswMannschaftLink:
            continue
        Ds = ProgMod_DsAngVer.calc(HeimMannschaftLink,AuswMannschaftLink)
        Ds_maxProg = Ds[Ds[:,2]==max(Ds[:,2])]
        DS_Prognose.append([HeimMannschaftLink, AuswMannschaftLink, Ds_maxProg])
```

```python
import ProgMod_GewDsAngVer
GewDS_Prognose = []
for HeimMannschaftLink in bl_teamslink:
    for AuswMannschaftLink in bl_teamslink:
        if HeimMannschaftLink == AuswMannschaftLink:
            continue
        GewDs = ProgMod_GewDsAngVer.calc(HeimMannschaftLink,AuswMannschaftLink)
        GewDs_maxProg = GewDs[GewDs[:,2]==max(GewDs[:,2])]
        GewDS_Prognose.append([HeimMannschaftLink, AuswMannschaftLink, GewDs_maxProg])
```

```python
testdaten = pd.read_csv("bl_mit_allspieler_2017.csv")
HeimID = testdaten["team1_id"]
AuswID = testdaten["team2_id"]
HeimTore = testdaten["tore_heim"]
AuswTore = testdaten["tore_aus"]
Ergebnisse = [HeimID,AuswID,HeimTore,AuswTore]
```

```python
#Ersetzt linknamen mit ID

for i in range(0,len(GewDS_Prognose)):
    
    if type(GewDS_Prognose[i][0]) == str:
        ProgHeimlink = GewDS_Prognose[i][0]
        GewDS_Prognose[i][0] = int(allTeams[allTeams['link']==ProgHeimlink]["team_id"])
        
    if type(GewDS_Prognose[i][1]) == str:
        ProgAuswlink = GewDS_Prognose[i][1]
        GewDS_Prognose[i][1] = int(allTeams[allTeams['link']==ProgAuswlink]["team_id"])
        
        
df=pd.DataFrame(GewDS_Prognose)
GewDS_PrognoseDF=df.rename(index=str, columns={0:"HeimID", 1: "AuswID",2: "GewDS_Prognose"})
```

```python
#Ersetzt linknamen mit ID

for i in range(0,len(DS_Prognose)):
    
    if type(DS_Prognose[i][0]) == str:
        ProgHeimlink = DS_Prognose[i][0]
        DS_Prognose[i][0] = int(allTeams[allTeams['link']==ProgHeimlink]["team_id"])
        
    if type(DS_Prognose[i][1]) == str:
        ProgAuswlink = DS_Prognose[i][1]
        DS_Prognose[i][1] = int(allTeams[allTeams['link']==ProgAuswlink]["team_id"])
        
        
df=pd.DataFrame(DS_Prognose)
DS_PrognoseDF=df.rename(index=str, columns={0:"HeimID", 1: "AuswID",2: "DS_Prognose"})
```

```python
ErgebnisVSGewDSProg=[]
for i in range(0,len(HeimID)):
    
    if sum(GewDS_PrognoseDF["HeimID"] == HeimID[i]) == 0:
        continue
    
    if sum(GewDS_PrognoseDF["AuswID"] == AuswID[i]) == 0:
        continue
    
    df1=GewDS_PrognoseDF[GewDS_PrognoseDF["HeimID"]==HeimID[i]]
    GewDS_ProgErg=df1[df1["AuswID"]==AuswID[i]]
    ErgebnisVSGewDSProg.append([HeimID[i],AuswID[i],HeimTore[i],AuswTore[i],GewDS_ProgErg["GewDS_Prognose"][0][0][0],GewDS_ProgErg["GewDS_Prognose"][0][0][1]])

df=pd.DataFrame(ErgebnisVSGewDSProg)
ErgebnisVSGewDSProgDF=df.rename(index=str, columns={0:"HeimID", 1: "AuswID",2: "HeimErg",3: "AuswErg",4: "HeimProg",5: "AuswProg"})
```

```python
ErgebnisVSDSProg=[]
for i in range(0,len(HeimID)):
    
    if sum(GewDS_PrognoseDF["HeimID"] == HeimID[i]) == 0:
        continue
    
    if sum(GewDS_PrognoseDF["AuswID"] == AuswID[i]) == 0:
        continue

    
    df1=DS_PrognoseDF[DS_PrognoseDF["HeimID"]==HeimID[i]]
    DS_ProgErg=df1[df1["AuswID"]==AuswID[i]]
    ErgebnisVSDSProg.append([HeimID[i],AuswID[i],HeimTore[i],AuswTore[i],DS_ProgErg["DS_Prognose"][0][0][0],DS_ProgErg["DS_Prognose"][0][0][1]])

df=pd.DataFrame(ErgebnisVSDSProg)
ErgebnisVSDSProgDF=df.rename(index=str, columns={0:"HeimID", 1: "AuswID",2: "HeimErg",3: "AuswErg",4: "HeimProg",5: "AuswProg"})
```

```python
GewDS_dataheim = np.array([ErgebnisVSGewDSProgDF["HeimErg"], ErgebnisVSGewDSProgDF["HeimProg"]])
GewDS_dataausw = np.array([ErgebnisVSGewDSProgDF["AuswErg"], ErgebnisVSGewDSProgDF["AuswProg"]])

print("### Gewichtetes Durchschnitts Modell ####")
print("Korrelationkoeffi für Heimergebnis " + str(round(np.corrcoef(GewDS_dataheim)[0][1],2)))
print("Korrelationkoeffi für Auswergebnis " + str(round(np.corrcoef(GewDS_dataausw)[0][1],2)))

print("Heimtore zu " + str(round(100*sum(ErgebnisVSGewDSProgDF["HeimErg"]==ErgebnisVSGewDSProgDF["HeimProg"])/len(ErgebnisVSGewDSProgDF["HeimErg"]),1),) + " % richtig getippt" )
print("Auswtore zu " + str(round(100*sum(ErgebnisVSGewDSProgDF["AuswErg"]==ErgebnisVSGewDSProgDF["AuswProg"])/len(ErgebnisVSGewDSProgDF["AuswErg"]),1),) + " % richtig getippt" )

A=ErgebnisVSGewDSProgDF["AuswErg"]==ErgebnisVSGewDSProgDF["AuswProg"] 
H=ErgebnisVSGewDSProgDF["HeimErg"]==ErgebnisVSGewDSProgDF["HeimProg"]
print("Ergebnis zu " + str(round(100*sum(A&H)/len(ErgebnisVSGewDSProgDF["AuswErg"]),2),) + " % richtig getippt" )
```
