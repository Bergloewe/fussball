```python
#Sucht linknamen der letzten Bundesligisten raus

allTeams = pd.read_csv("TeamID_Team.csv")
bl_TeamsID = set(pd.read_csv("bl_mit_allspieler.csv")["team1_id"][-200:])

bl_teamslink=[]
for teamID in bl_TeamsID:
    bl_teamslink.append(allTeams['link'][teamID-1])
```

```python
import ProgMod_DsAngVer
DS_Prognose = []
for HeimMannschaftLink in bl_teamslink:
    for AuswMannschaftLink in bl_teamslink:
        if HeimMannschaftLink == AuswMannschaftLink:
            continue
        Ds = ProgMod_DsAngVer.calc(HeimMannschaftLink,AuswMannschaftLink)
        Ds_maxProg = Ds[Ds[:,2]==max(Ds[:,2])]
        DS_Prognose.append([HeimMannschaftLink, AuswMannschaftLink, Ds_maxProg])
```

```python
import ProgMod_GewDsAngVer
GewDS_Prognose = []
for HeimMannschaftLink in bl_teamslink:
    for AuswMannschaftLink in bl_teamslink:
        if HeimMannschaftLink == AuswMannschaftLink:
            continue
        GewDs = ProgMod_GewDsAngVer.calc(HeimMannschaftLink,AuswMannschaftLink)
        GewDs_maxProg = GewDs[GewDs[:,2]==max(GewDs[:,2])]
        GewDS_Prognose.append([HeimMannschaftLink, AuswMannschaftLink, GewDs_maxProg])
```

```python
testdaten = pd.read_csv("bl_mit_allspieler_2016.csv")
HeimID = testdaten["team1_id"]
AuswID = testdaten["team2_id"]
HeimTore = testdaten["tore_heim"]
AuswTore = testdaten["tore_aus"]
Ergebnisse = [HeimID,AuswID,HeimTore,AuswTore]
```

```python
#Ersetzt linknamen mit ID

for i in range(0,len(GewDS_Prognose)):
    
    if type(GewDS_Prognose[i][0]) == str:
        ProgHeimlink = GewDS_Prognose[i][0]
        GewDS_Prognose[i][0] = int(allTeams[allTeams['link']==ProgHeimlink]["team_id"])
        
    if type(GewDS_Prognose[i][1]) == str:
        ProgAuswlink = GewDS_Prognose[i][1]
        GewDS_Prognose[i][1] = int(allTeams[allTeams['link']==ProgAuswlink]["team_id"])
        
        
df=pd.DataFrame(GewDS_Prognose)
renamed=df.rename(index=str, columns={0:"HeimID", 1: "AuswID",2: "Prognose"})
GewDS_Prognose=renamed
```

```python
df1=GewDS_Prognose[GewDS_Prognose["HeimID"]==HeimID[0]]
df2=df1[df1["AuswID"]==AuswID[0]]
```

```python
df1=GewDS_Prognose[GewDS_Prognose["HeimID"]==HeimID[0]]
ProgErg=df1[df1["AuswID"]==AuswID[0]]
ProgErg["Prognose"]
```

```python
ErgebnisVSProg=[]
for i in range(0,len(HeimID)):
    
    if HeimID[i]==46 or HeimID[i]==43:
        continue
    
    if AuswID[i]==46 or AuswID[i]==43:
        continue
    
    df1=GewDS_Prognose[GewDS_Prognose["HeimID"]==HeimID[i]]
    ProgErg=df1[df1["AuswID"]==AuswID[i]]
    ErgebnisVSProg.append([HeimID[i],AuswID[i],HeimTore[i],AuswTore[i],ProgErg["Prognose"][0][0][0],ProgErg["Prognose"][0][0][1]])

df=pd.DataFrame(ErgebnisVSProg)
ErgebnisVSProg=df.rename(index=str, columns={0:"HeimID", 1: "AuswID",2: "HeimErg",3: "AuswErg",4: "HeimProg",5: "AuswProg"})
```

```python
import matplotlib.pyplot as plt
plt.plot(ErgebnisVSProg["HeimErg"], ErgebnisVSProg["HeimProg"], 'ro')
plt.axis([0, 10, 0, 3])
plt.xlabel("HeimErg")
plt.ylabel("HeimProg")
plt.show()
```

```python
dataheim = np.array([ErgebnisVSProg["HeimErg"], ErgebnisVSProg["HeimProg"]])
print("Der Korrelationkoeffi für das Heimergebnis liegt bei " + str(np.corrcoef(dataheim)[0][1]))
print("Das Model hat dei Heimtore zu " + str(round(100*sum(ErgebnisVSProg["HeimErg"]==ErgebnisVSProg["HeimProg"])/len(ErgebnisVSProg["HeimErg"])),) + " % richtig getippt" )
```

```python
dataausw = np.array([ErgebnisVSProg["AuswErg"], ErgebnisVSProg["AuswProg"]])
print("Der Korrelationkoeffi für das Auswergebnis liegt bei " + str(np.corrcoef(dataausw)[0][1]))
print("Das Model hat dei Auswtore zu " + str(round(100*sum(ErgebnisVSProg["AuswErg"]==ErgebnisVSProg["AuswProg"])/len(ErgebnisVSProg["AuswErg"])),) + " % richtig getippt" )
```

```python
A=ErgebnisVSProg["AuswErg"]==ErgebnisVSProg["AuswProg"] 
H=ErgebnisVSProg["HeimErg"]==ErgebnisVSProg["HeimProg"]
print("Das Model hat dei Auswtore zu " + str(round(100*sum(A*H)/len(ErgebnisVSProg["AuswErg"])),) + " % richtig getippt" )
```
